## 1. Definición de su modelo cosmológico

Una de las ideas cruciales que aprenderán en el curso, es que todo está sujeto a su modelo cosmológico.

En este caso, consideremos los parámetros de [Planck](https://arxiv.org/abs/1807.06209):

| | |
| - | - |
| $H_0 = (67.4 ± 0.5) km/s/Mpc $ | Hubble constant |
| $\Omega_m = 0.315 ± 0.0071$ | matter density parameter |
| $\Omega_c h^2: 0.120 ± 0.0011$ | dark matter density |
| $\Omega_b h^2 = 0.0224 ± 0.00011$ | baryon density |
| $n_s = 0.965 ± 0.0041$ | scalar spectral index | 
| $\tau = 0.054 ± 0.0071$ | optical depth | 

```python
H0 = 67.4 # km/s/Mpc
omega_m = 0.315
omega_ch2 = 0.120
omega_bh2 = 0.0224
ns = 0.965
tau = 0.054
```

Debemos darle un modelo cosmológico a CAMB, para que podamos hacer predicciones. Tenemos dos formas: `camb.CAMBparams()` y `camb.set_params()`.

### 1.1 `camb.CAMBparams()`

```python
help(camb.CAMBparams)
```

```python
# Creamos un objeto donde guardaremos los parámetros cosmológicos
pars = camb.CAMBparams()

pars.set_cosmology(H0=H0, ombh2=omega_bh2, omch2=omega_ch2, tau=tau)
```

```python
# Podemos cambiar algún parámetro después de fijar el modelo.

pars.set_dark_energy(w=-1);
# Nos dice que la energía oscura es una constante (Lambda CDM)
# \dot{\rho} + 3 H (1+ w) \rho =0 

pars
```

```python
# Ahora, para obtener resultados podemos simplemente

results = camb.get_results(pars)
```

### 1.2 `camb.set_params()`

```python
pars2 = camb.set_params(H0=67.4, w=-0.1, cs2=0.1)
results2 = camb.get_results(pars2)
```

#### **Hágalo usted mismx:**


Comparar con la solución analítica con la de CAMB

$$ D_a(z) = \frac{f_K(\chi)}{1+z}, \quad D_L(z) = (1+z)f_K(\chi)$$ 

Donde:
\begin{align}
    f_K(\chi) =
    \begin{dcases}
        \frac{c}{H_0\sqrt{\Omega_{K,0}}} \sinh{\left(\sqrt{\Omega_{K,0}} \int_0^z \frac{dx}{E(x)}\right)} & \Omega_{k,0}>0 \\
        \frac{c}{H_0}\int_0^z \frac{dx}{E(x)} & \Omega_{k,0}=0\\
        \frac{c}{H_0\sqrt{-\Omega_{K,0}}} \sin{\left(\sqrt{-\Omega_{K,0}} \int_0^z \frac{dx}{E(x)}\right)} & \Omega_{k,0}<0\\
    \end{dcases}
\end{align}


\begin{align}
    E(z) = H(z) / H_0
\end{align}


\begin{align}
    H^2(z) = H_0^2 \left[\Omega_{r,0}(1+z)^4 + \Omega_{m,0}(1+z)^3 + \Omega_{\Lambda,0}(1+z)^{3(1+w)} + \Omega_{K,0}(1+z)^2\right]
\end{align}

```python
w = -1
c = 3 * 10**5

Omega_k = 0
Omega_r = 10**(-4)
Omega_lambda = 0.75
Omega_c = 0.2
Omega_b = 0.05
Omega_m = Omega_b + Omega_c

def Hubble2(z, H0, Omega_r ,Omega_m, Omega_lambda, Omega_k, w):
    return H0**2 * (Omega_r*(1+z)**4 + Omega_m*(1+z)**3 +  Omega_lambda*(1+z)**(3*(1+w)) + Omega_k*(1+z)**2)
```

```python
fig, ax = plt.subplots(ncols=1, nrows=1, sharey=True, figsize=(6, 4), constrained_layout = True)

ax.loglog(zs, Hubble2(zs, H0, Omega_r ,Omega_m, Omega_lambda, Omega_k, w), color='purple', label='Hubble parameter')

ax.grid(alpha=0.4, ls=':')

ax.legend()
ax.set_xlabel(r'Redshift $z$', fontsize=16)
ax.set_ylabel(r'$H(z)^2$ [Mpc]', fontsize=16)
```

```python

def E(z, H0, Omega_r ,Omega_m, Omega_lambda, Omega_k, w):
    return np.sqrt(Hubble2(z, H0, Omega_r ,Omega_m, Omega_lambda, Omega_k, w))/ H0

def func(z, H0, Omega_r ,Omega_m, Omega_lambda, Omega_k, w):
    return 1/E(z, H0, Omega_r ,Omega_m, Omega_lambda, Omega_k, w)

def f_k(z, H0, Omega_r ,Omega_m, Omega_lambda, Omega_k, w):
    if Omega_k > 0:
        return (c/(H0*np.sqrt(Omega_k))) *np.sinh(np.sqrt(Omega_k)* quad(func, 0, z, args=(H0, Omega_r ,Omega_m, Omega_lambda, Omega_k, w)))[0]
    elif Omega_k == 0:
        return (c/H0) * quad(func, 0, z, args=(H0, Omega_r ,Omega_m, Omega_lambda, Omega_k, w))[0]
    elif Omega_k < 0:
        return (c/(H0*np.sqrt(-Omega_k))) *np.sin(np.sqrt(-Omega_k) * quad(func, 0, z, args=(H0, Omega_r ,Omega_m, Omega_lambda, Omega_k, w)))[0]

def D_A(z, H0, Omega_r ,Omega_m, Omega_lambda, Omega_k, w): 
    return f_k(z, H0, Omega_r ,Omega_m, Omega_lambda, Omega_k, w)/ (1+z)

def D_L(z, H0, Omega_r ,Omega_m, Omega_lambda, Omega_k, w):
    return (1+z) * f_k(z, H0, Omega_r ,Omega_m, Omega_lambda, Omega_k, w)
```

```python
fig, ax = plt.subplots(ncols=1, nrows=1, sharey=True, figsize=(6, 4), constrained_layout = True)


DA_values = [D_A(z, H0, Omega_r, Omega_m, Omega_lambda, Omega_k, w) for z in zs]
DL_values = [D_L(z, H0, Omega_r, Omega_m, Omega_lambda, Omega_k, w) for z in zs]

ax.loglog(zs, D_ang, color='purple', label='Angular Diameter Distance')
ax.loglog(zs, D_lum, color='b', label='Luminosity Distance')
ax.loglog(zs, DA_values, '--', label='D_A solución analítica')
ax.loglog(zs, DL_values, '--',label='D_L solución analítica')

ax.grid(alpha=0.4, ls=':')

ax.legend()
ax.set_xlabel(r'Redshift $z$', fontsize=16)
ax.set_ylabel(r'$Distances$ [Mpc]', fontsize=16)
```
